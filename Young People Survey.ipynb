```python
from sklearn.model_selection import train_test_split

# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

import numpy as np
from sklearn.metrics import accuracy_score
```

```python
target = data_copy5['HI']
x_train, x_test, y_train, y_test = train_test_split(data_copy5, target, random_state = 42)
```

```python
target_drop = ['HI']
x_train = x_train.drop(target_drop, axis = 1)
x_test = x_test.drop(target_drop, axis = 1)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
k_fold = KFold(n_splits=10, shuffle = True, random_state = 0)
```

```python
def modeltraining(clf):
    scoring = 'accuracy'
    score = cross_val_score(clf, x_train, y_train, cv = k_fold, n_jobs = 1, scoring = scoring)
    print(round(np.mean(score)*100, 2))
```

```python
modeltraining(KNeighborsClassifier(n_neighbors = 13))
modeltraining(DecisionTreeClassifier())
modeltraining(RandomForestClassifier(n_estimators=13))
modeltraining(GaussianNB())
modeltraining(SVC(gamma=0.001))
modeltraining(LogisticRegression(solver = 'lbfgs', multi_class = 'multinomial', random_state = 42))
#modeltraining(LinearRegression())
```

```python
def svmScore(c = 1, gam = 0.001, ker ='rbf'):
    svMachine = SVC(C = c, kernel = ker, gamma = gam)
    svMachine.fit(x_train, y_train)
    y_pred = svMachine.predict(x_test)
    print(accuracy_score(y_test, y_pred))
```

```python
# compare different kernels
# the gamma is default to be 1 / n_features
kernels = ['linear', 'rbf', 'poly']
for k in kernels:
    svmScore(ker = k)
```

```python
#Random Forest
n_trees = np.arange(1, 20, 4) # build a array contains values form 1 to 100 with step of 4

# for each n, we want to check the accuracy of random forest with n trees
for n in n_trees:
    rdf = RandomForestClassifier(n_estimators = n, random_state = 42)
    rdf.fit(x_train, y_train)
    # we will print both the number of trees, and the accuracy score
    print(n)
    print(accuracy_score(y_test, rdf.predict(x_test)))
```

```python
#Hierarchical Clustering
from sklearn import datasets
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# generate the linkage matrix
Z = linkage(data_copy6, 'ward')

# set cut-off to 150 cluster merges
max_d = 7.08                # max_d as in max_distance

plt.figure(figsize=(25, 10))
plt.title('Music Preference Groups')
plt.xlabel('Music Group')
plt.ylabel('distance')
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=150,                  # Try changing values of p
    leaf_rotation=90.,      # rotates the x axis labels
    leaf_font_size=8.,      # font size for the x axis labels
)
plt.axhline(y=max_d, c='k')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(linkage="ward", n_clusters=4)
clustering.fit(data_copy6);
```

```python
data_spending_copy_binned.head()
```

```python
def bar_chart(feature):
    Low_Spend = data_spending_copy_binned[data_spending_copy_binned['Spending Group']=='Low Spend'][feature].value_counts()
    Med_Spend = data_spending_copy_binned[data_spending_copy_binned['Spending Group']=='Med Spend'][feature].value_counts()
    High_Spend = data_spending_copy_binned[data_spending_copy_binned['Spending Group']=='High Spend'][feature].value_counts()
    Very_High_Spend = data_spending_copy_binned[data_spending_copy_binned['Spending Group']=='Very High Spend'][feature].value_counts()
    df = pd.DataFrame([Low_Spend,Med_Spend,High_Spend,Very_High_Spend, ((High_Spend + Very_High_Spend)/(Low_Spend+Med_Spend+High_Spend+Very_High_Spend))*100])
    df.index = ['Low_Spend','Med_Spend','High_Spend','Very_High_Spend',"Proportation of high spenders"]
    print(df)
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
is2 = data_copy6['labels'] == 2
data_copy7 = data_copy6[is2]
data_copy8 = data_copy7.drop(columns=['Music'])
data_copy8.head()
```

```python
def clusterCount(label):
    isLabel = data_copy6['labels'] == label
    data_copy7 = data_copy6[isLabel]
    data_copy8 = data_copy7.drop(columns=['Music'])
    
    columns = ['genre','average']
    index = range(0,len(data_copy8.columns))
    
    label2 = pd.DataFrame(index=index, columns=columns)
    label2['genre'] = list(data_copy8.columns.values)
    label2['average'] = list(data_copy8.mean(axis=0).values)

    isMax = label2['average'] == data_copy8.mean(axis=0).max()
    label2Max = label2[isMax]
    print(label2Max.head())
```

```python
clusterCount(0)
clusterCount(1)
clusterCount(2)
clusterCount(3)
```

From this analysis, we can conclude that the highest spending music group is the cluster 2, which has highest preference to Rock music
